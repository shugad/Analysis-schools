## Выгрузка данных ## 

Объединяем все предыдущие шаги в одном коде. 

```
contracts_tab = [] #в этот список будем складывать информацию о каждом контракте
for page in range(1, 8):
  response = requests.get(f'https://api.spending.gov.ru/v1/natprojects/contracts?fp_id=E4&page={page}')
  print(page)
  print(response)
  contracts = response.json()
  lst = contracts['items'] # список контрактов на одной странице
  for cnt in lst: # проход по каждому контракту
    list1=cnt['fp_info'] # лист со словарем, содержащим основную информацию о сумме контракта
    a=list1[0] # доступ к словарю
    one_cont = [] # сюда складывается информация об одном контракте
    one_cont.append(cnt['customer_name']) # наименование заказчика
    #one_cont.append(cnt['customer_inn'])
    #one_cont.append(cnt['customer_kpp'])
    one_cont.append(a['fp_price']) # сумма контракта
    list2=cnt['suppliers'] # лист со словарем, содержащим основную информацию о поставщике
    b=list2[0] # доступ к словарю
    one_cont.append(b['supplier_name']) # добавляем инфо по одному контракту
    one_cont.append(cnt['year'])
    one_cont.append(cnt['regnum'])
    contracts_tab.append(one_cont) # сюда складывается информация о всех контрактах
```

```
# проверим выгруженные значения в итоговом списке

contracts_tab
```

```
# с помощью метода pd.DataFrame преобразуем список в датафрейм (таблицу)

frame = pd.DataFrame(contracts_tab)
frame
```

```
#зададим названия столбцам

frame.rename(columns={0: 'customer', 1: 'value', 2: 'supplier', 3: 'year', 4: 'regnum'}, inplace=True) 
```

```
#посмотрим, что получилось

frame = frame.astype({"regnum": 'int64'}) # меняем тип столбика с номером контракта на int64 
frame.head()
```

# 2. Выгрузка кодов регионов по номерам контрактов
Чтобы сопоставить контракты и их регионы для построения интерактивной карты, нужно получить дополнительную информацию с карточки конкретного контракта.

```
response = requests.get(f'https://api.spending.gov.ru/v1/contracts/search/?regnum=2890102552719000007') # поиск по номеру контракта 2890102552719000007
print(response)
contracts = response.json()
regionCode = contracts['contracts']['data'][0]['regionCode'] # структура хранения кода региона в json файле
regionCode
```

```
reg_and_number_all = [] # сюда складывается вся информация по всем конрактам
for reg_num in frame.regnum: # проходим по всем номерам имеющихся контрактов
  reg_and_number = [] # сюда складывается вся информация по одному контракту
  response = requests.get(f'https://api.spending.gov.ru/v1/contracts/search/?regnum={reg_num}') # подставляем в запрос номер
  print(reg_num)
  print(response)
  contracts = response.json()
  regionCode = contracts['contracts']['data'][0]['regionCode'] # получаем код региона
  reg_and_number.append(reg_num) # добавляем номер контракта для идентификации
  reg_and_number.append(regionCode) # добавляем код региона
  reg_and_number_all.append(reg_and_number) 
```

```
reg_and_number_all
```

```
df_reg = pd.DataFrame(reg_and_number_all, columns =['regnum', 'id_region']) # сохраняем выгрузку в датафрейм
df_reg = df_reg.astype('int64') # меняем тип всех столбиков на int64
```

```
full = frame.merge(df_reg, on='regnum', how='left') # присоединяем к выгрузке всех контрактов столбик с кодами регионов по общему столбику "номеру контракта"
full
```
